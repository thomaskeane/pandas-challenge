```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file_to_load = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
raw_df = pd.read_csv(file_to_load)
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
df_reduced = raw_df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
reduced_cols = df_reduced.columns.tolist()
reduced_cols
```

```python
# Change reading 0 and 1 to No and Yes, respectively
df_reduced = df_reduced.replace([0.0],'No')
df_reduced = df_reduced.replace([1.0], 'Yes')
df_reduced.head(50)
```

```python
# Extract rows for only those who attended a bootcamp
attended = df_reduced.loc[df_reduced['AttendedBootcamp']=='Yes']
attended.head(20)
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
attended_clean = attended.dropna(subset = ['BootcampName'])
boot_sig = attended_clean[['BootcampName', 'AttendedBootcamp']].groupby('BootcampName').count()\
        .rename(columns={'AttendedBootcamp':'Attended'}).sort_values('Attended', ascending = False)
#gender_breakdown
boot_sig
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
attend_yes = attended_clean.loc[attended_clean['BootcampRecommend']=='Yes']
boot_rec = attend_yes[['BootcampName', 'BootcampRecommend']].groupby('BootcampName').count()\
            .rename(columns={'BootcampRecommend':'Recommended'})\
            .sort_values('Recommended', ascending = False)
boot_rec
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merge_df = pd.merge(boot_sig, boot_rec, on="BootcampName", how="outer")
merge_df.head(20)
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
merge_df['Percent'] = merge_df[['Recommended']].div(merge_df.Attended, axis=0)
merge_df['Recommended'] = merge_df.Recommended.map("{:.0f}".format)
merge_df['Percent'] = merge_df.Percent.map("{:.2%}".format)
```

```python
# Export to excel and remove index
outtable = merge_df.reset_index()
outtable
```

```python
outtable.to_excel(r'BootcampOutputPart2.xlsx', index = False)
```
