```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.dtypes
```

```python
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Alt: tot_players = purchase_data['SN'].nunique()
unique_players = purchase_data.drop_duplicates(subset='SN', keep='first')
tot_players = len(unique_players)
tot_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
nu_items = purchase_data['Item Name'].nunique()
ap_items = "${:,.2f}".format(purchase_data['Price'].mean())
purchases = purchase_data['Purchase ID'].count()
tot_rev = "${:,.2f}".format(purchase_data['Price'].sum())
```

```python
summary_df = pd.DataFrame(
            {"Number of Unique Items": [nu_items],
             "Average Price": [ap_items],
             "Number of Purchases": [purchases],
             "Total Revenues": [tot_rev]})
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Alt:gender_demo = purchase_data[['SN','Gender']].drop_duplicates(subset='SN', keep='first').groupby('Gender').count()
gender_demo = unique_players[['SN','Gender']].groupby('Gender').count().rename(columns={'SN':"Total Count"})
gender_demo['Percentage of Players']= (gender_demo['Total Count'] / tot_players).map("{:.2%}".format)
gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Calculations
gender_pa = purchase_data[['Purchase ID','Gender']].groupby('Gender').count()\
            .rename(columns={'Purchase ID': 'Purchase Count'})
gender_pa['Average Purchase Price']= purchase_data[['Price','Gender']].groupby('Gender').mean()
gender_pa['Total Purchase Value'] = purchase_data[['Price','Gender']].groupby('Gender').sum()
gender_pa['Avg Total Purchase per Person'] = purchase_data[['SN','Price','Gender']].groupby(['SN', 'Gender']).sum()\
                                             .groupby('Gender').mean()

# Formatting
gender_pa['Average Purchase Price'] = gender_pa['Average Purchase Price'].map("${:.2f}".format)
gender_pa['Total Purchase Value'] = gender_pa['Total Purchase Value'].map("${:.2f}".format)
gender_pa['Avg Total Purchase per Person'] = gender_pa['Avg Total Purchase per Person'].map("${:.2f}".format)

gender_pa
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Bins for the following exercises
bins = [0,9,14,19,24,29,34,39,1000]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
```

```python
# Player counts: have to use unique players so that we count players only once
unique_players['Age Ranges'] = pd.cut(unique_players.Age, bins, labels=group_names)
age_count = unique_players[['SN','Age Ranges']].groupby('Age Ranges').count().rename(columns={'SN':"Total Count"})
age_count['Percentage of Players']= (age_count['Total Count'] / tot_players).map("{:.2%}".format)
age_count
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Calculations
purchase_data['Age Ranges'] = pd.cut(purchase_data.Age, bins, labels=group_names)

age_pa = purchase_data[['Purchase ID','Age Ranges']].groupby('Age Ranges').count()\
            .rename(columns={'Purchase ID': 'Purchase Count'})
age_pa['Average Purchase Price']= purchase_data[['Price','Age Ranges']].groupby('Age Ranges').mean()
age_pa['Total Purchase Value'] = purchase_data[['Price','Age Ranges']].groupby('Age Ranges').sum()
age_pa['Avg Total Purchase per Person'] = purchase_data[['SN','Price','Age Ranges']].groupby(['SN', 'Age Ranges']).sum()\
                                             .groupby('Age Ranges').mean()

# Formatting
age_pa['Average Purchase Price'] = age_pa['Average Purchase Price'].map("${:.2f}".format)
age_pa['Total Purchase Value'] = age_pa['Total Purchase Value'].map("${:.2f}".format)
age_pa['Avg Total Purchase per Person'] = age_pa['Avg Total Purchase per Person'].map("${:.2f}".format)
age_pa
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Calculations
ts = purchase_data[['SN','Price']].groupby(['SN']).sum()\
    .rename(columns={'Price': 'Total Purchase Value'})\
    .sort_values('Total Purchase Value', ascending=False)
ts['Purchase Count'] = purchase_data[['Purchase ID','SN']].groupby('SN').count()
ts['Average Purchase Price'] = purchase_data[['SN', 'Price']].groupby('SN').mean()


# Formatting
ts['Total Purchase Value'] = ts['Total Purchase Value'].map("${:.2f}".format)
ts['Average Purchase Price'] = ts['Average Purchase Price'].map("${:.2f}".format)
ts = ts[['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']]
ts.head()
```
