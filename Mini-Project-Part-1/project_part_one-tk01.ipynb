```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file_to_load = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
df = pd.read_csv(file_to_load)
```

```python
# Inspect all columns
df.head()
```

```python
cols_all = df.columns.tolist()
cols_all
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
df_reduced = df.iloc[:, [0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
cols_reduced = df_reduced.columns.tolist()
cols_reduced
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
df_reduced = df_reduced.replace([0.0],'No')
df_reduced = df_reduced.replace([1.0], 'Yes')
df_reduced.head(50)
```

```python
# Calculate total number of respondents in survey
nr = len(df_reduced.index)
nr
```

```python
# Extract rows corresponding only to people who attended a bootcamp
df_attended = df_reduced.loc[df_reduced['AttendedBootcamp']=='Yes']
df_attended.head(100)
```

```python
# Calculate average age of attendees
mean_attendee_age = df_attended['Age'].mean()
mean_attendee_age
```

```python
# Calculate how many people attended a bootcamp
number_attended_bootcamp = len(df_attended.index)
number_attended_bootcamp
```

```python
# Calculate how many attendees hold degrees
#hold_degree = df_attended["SchoolDegree"].count()
#hold_degree
hold_degree = number_attended_bootcamp - df_attended['SchoolMajor'].isnull().sum()
hold_degree
```
