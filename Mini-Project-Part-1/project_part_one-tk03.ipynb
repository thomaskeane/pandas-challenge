```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file_to_load = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
raw_df = pd.read_csv(file_to_load)
```

```python
# Inspect all columns
raw_df.head()
```

```python
raw_cols = raw_df.columns.tolist()
raw_cols
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
df_reduced = raw_df.iloc[:, [0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
reduced_cols = df_reduced.columns.tolist()
reduced_cols
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
df_reduced = df_reduced.replace([0.0],'No')
df_reduced = df_reduced.replace([1.0], 'Yes')
df_reduced.head(50)
```

```python
# Calculate total number of respondents in survey
num_respondents = len(df_reduced.index)
num_respondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended = df_reduced.loc[df_reduced['AttendedBootcamp']=='Yes']
attended.head(20)
```

```python
# Calculate average age of attendees
mean_attendee_age = "{:.1f}".format(attended['Age'].mean())
mean_attendee_age
```

```python
# Calculate how many people attended a bootcamp
number_attended_bootcamp = len(attended.index)
number_attended_bootcamp
```

```python
#Calculate how many attendees hold degrees
hold_degree = number_attended_bootcamp - attended['SchoolMajor'].isnull().sum()
hold_degree
```

```python
#Count number of attendees who self-identify as male; female; or are of non-binary gender identification
gender_breakdown = attended[['AttendedBootcamp', 'Gender']].groupby('Gender').count()\
                 .rename(columns={'AttendedBootcamp':'Count'})
gender_breakdown
```

```python
# Calculate percentage of respondents who attended a bootcamp
percent_attended = "{:.2%}".format(number_attended_bootcamp / num_respondents)
percent_attended
```

```python
# Calculate percentage of respondents belonging to each gender
gender_breakdown['Percent'] = (gender_breakdown['Count'] / gender_breakdown.Count.sum())
#gender_breakdown.reset_index(inplace=True)
non_binary = "{:.2%}".format(gender_breakdown.loc['agender'].Percent+gender_breakdown.loc['genderqueer'].Percent+gender_breakdown.loc['trans'].Percent)
gender_breakdown['Percent'] = gender_breakdown.Percent.map("{:.2%}".format)
gender_breakdown

```

```python
non_binary
```

```python
# Calculate percentage of attendees with a school degree
percent_degree = "{:.2%}".format(hold_degree / number_attended_bootcamp)
percent_degree
```

```python
# Calculate average post-bootcamp salary
avg_post_salary = "${:,.2f}".format(attended['BootcampPostSalary'].mean())
avg_post_salary
```

```python
# Create a new table consolodating above calculations
outtable = pd.DataFrame({'Total Surveyed':[num_respondents],
                         'Total Bootcamp attendees':[number_attended_bootcamp],
                         '% attended bootcamp':[percent_attended],
                         'Avg. Age':[mean_attendee_age],
                         'Has a degree':[percent_degree],
                         '% Male':[gender_breakdown.loc['male'].Percent],
                         '% Female':[gender_breakdown.loc['female'].Percent]
                         '% Non-Binary':[gender]
                        })
outtable
```
